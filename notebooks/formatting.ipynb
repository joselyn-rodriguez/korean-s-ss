### Formatting

Quick code to help annotate TextGrids faster. It does the following things:

1. Deletes unnecessary tiers for annotation
2. Adds a new tier for feature
3. Adds frication and aspiration in the feature tier in the position of the s0 or ss phoneme
4. Saves the file in the same folder it was found in

```python
from praatio import textgrid

tg_list = get_filelist(Path.cwd(), "TextGrid")

for i in tg_list:
    print(i)
    tg = textgrid.openTextgrid(i, includeEmptyIntervals=True, duplicateNamesMode='rename')
    tiers = tg.tierNames
    keep_tiers = {'phoneme','pWord.ortho.', 'pWord.ortho._2'}

    for tier_name in tiers:
        if tier_name not in keep_tiers:
            tg.removeTier(tier_name)

    phoneme_tier = tg.getTier('phoneme')

    s_intervals = [
        interval for interval in phoneme_tier.entries
        if interval.label == 'ss'
    ]

    s = s_intervals[0]

    if s.start < 0.21:
        s = s_intervals[1]

    start = s.start
    end = s.end

    # example: split s0 into two features ðŸ˜„
    mid = start + (end - start) / 2

    new_tier = textgrid.IntervalTier(
        name="feature",
        entries=[(start, mid, 'frication'), (mid, end, 'aspiration')],
        minT=tg.minTimestamp,
        maxT=tg.maxTimestamp
    )

    tg.addTier(new_tier, tierIndex=1)
    tg.save(i, format='short_textgrid', includeBlankSpaces=True)
```
