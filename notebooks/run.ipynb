### Count word-initial ㅅ
Count instances of ㅅ at the beginning of the word (pronounced)

```python
filename1 = 'CSV/ej_pron.csv'
filename2 = 'CSV/participants2040pron.csv'

print("s0", count(filename1, '^s0'))

for vowel in vowels:
    searchStr = '^s0' + vowel
    print("s0" + vowel, count(filename1, searchStr))

print("s0 between 20-50", count(filename2, '^s0oo'))
```

Count instances of ㅅ at the beginning of the word (orthographic)

```python
filename1 = 'CSV/ej_orth.csv'
filename2 = 'CSV/participants2040ortho.csv'

print("s0", count(filename1, '^s0'))

for vowel in vowels:
    searchStr = '^s0' + vowel
    print("s0" + vowel, count(filename1, searchStr))

print("s0 between 20-50", count(filename2, '^s0oo'))
```

### Count word-initial ㅆ
Count instances of ㅆ at the beginning of the word (pronounced)

```python
filename1 = 'CSV/ej_pron.csv'
filename2 = 'CSV/participants2040pron.csv'

print("ss", count(filename1, '^ss'))

for vowel in vowels:
    searchStr = '^ss' + vowel
    print("ss" + vowel, count(filename1, searchStr))

print("ss between 20-50", count(filename2, '^ssoo'))
```

Count instances of ㅆ at the beginning of the word (orthographic)

```python
filename1 = 'CSV/ej_orth.csv'
filename2 = 'CSV/participants2040ortho.csv'

print("ss", count(filename1, '^ss'))

for vowel in vowels:
    searchStr = '^ss' + vowel
    print("ss" + vowel, count(filename1, searchStr))

print("ss between 20-50", count(filename2, '^ssoo'))
```

### Count word-medial ㅅ and ㅆ

Count instances of ㅅ and ㅆ in the middle of the word (pronounced)

```python
filename = 'CSV/ej_pron.csv'
V = '.[Eaeiouvx]'

print("s0", count(filename, V+'(-)?s0(-)?'+V))
print("ss", count(filename, V+'(-)?ss(-)?'+V))
```

```python
filename20 = 'CSV/participants20pron.csv'
filename30 = 'CSV/participants30pron.csv'
filename40 = 'CSV/participants40pron.csv'
filename20_40 = 'CSV/participants2040pron.csv'

V = '.[Eaeiouvx]'

print("Word-medial s0oo")
print("Vs0oo 20s", count(filename20, V+'(-)?s0-?oo'))
print("Vs0oo 30s", count(filename30, V+'(-)?s0-?oo'))
print("Vs0oo 40s", count(filename40, V+'(-)?s0-?oo'))
print("Vs0oo 20s-40s", count(filename20_40, V+'(-)?s0-?oo'))

print("\nWord-medial ssoo")
print("Vssoo 20s", count(filename20, V+'(-)?ss-?oo'))
print("Vssoo 30s", count(filename30, V+'(-)?ss-?oo'))
print("Vssoo 40s", count(filename40, V+'(-)?ss-?oo'))
print("Vssoo 20s-40s", count(filename20_40, V+'(-)?ss-?oo'))
```
